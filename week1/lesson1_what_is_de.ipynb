## 1.1 The Role of a Data Engineer

A **data engineer** builds and maintains the infrastructure that makes data usable.
Their output is reliable, clean, and accessible data — the fuel that analysts and data scientists run on.

Think of it like a city's water system. Data engineers build the pipes, pumps, and treatment plants.
Analysts and scientists are the end users who turn on the tap.

### Core responsibilities

| Responsibility | Description | Example Tools |
|----------------|-------------|---------------|
| **Ingest** | Pull data from source systems (APIs, databases, files, streams) | Fivetran, Airbyte, Kafka, Python scripts |
| **Store** | Land data in a data warehouse, data lake, or database | Snowflake, BigQuery, S3, PostgreSQL |
| **Transform** | Clean, normalize, and model data into usable structures | dbt, Spark, pandas |
| **Orchestrate** | Schedule and monitor pipelines to keep data flowing reliably | Airflow, Prefect, Dagster |
| **Serve** | Expose data to consumers (BI tools, ML models, apps) | Looker, Tableau, REST APIs |

## 1.2 Data Engineering vs Data Analysis vs Data Science

All three roles work with data, but they answer different questions.

```
Raw Data
   │
   ▼
[Data Engineer] ──── builds pipelines, stores & transforms data
   │
   ▼
Clean, Structured Data
   │
   ├──▶ [Data Analyst]    ──── asks "What happened?" → SQL, BI dashboards
   │
   └──▶ [Data Scientist]  ──── asks "What will happen?" → ML models, statistics
```

> **Data engineers are the prerequisite for everything else.**

| Dimension | Data Engineer | Data Analyst | Data Scientist |
|-----------|--------------|-------------|----------------|
| **Primary question** | How do we get the data? | What does the data say? | What will the data predict? |
| **Core skills** | SQL, Python, pipelines, cloud | SQL, Excel, visualization | Python/R, statistics, ML |
| **Output** | Pipelines & data models | Reports & dashboards | Models & predictions |
| **Example tools** | Spark, dbt, Airflow, Kafka | Tableau, Looker, Power BI | scikit-learn, TensorFlow, notebooks |
| **Focus** | Reliability & scale | Business insight | Prediction & experimentation |

## 1.3 What Data Engineers Actually Build

### ETL vs ELT

**ETL** (Extract, Transform, Load) — older approach: transform data *before* loading it into storage.

**ELT** (Extract, Load, Transform) — modern approach: load raw data first, then transform *inside* the warehouse.
ELT dominates today because cloud warehouses are cheap and powerful enough to run transforms at scale.

```
Source Systems          Storage              Consumers
──────────────          ───────              ─────────
Postgres DB   ──┐
Salesforce    ──┼──► Data Warehouse ──────► BI Tool (Looker)
REST API      ──┤       (Snowflake)         ML Model
Kafka stream  ──┘                           Downstream App
```

### Storage types

| Concept | Description | Example Tools |
|---------|-------------|---------------|
| **Data Warehouse** | Structured, query-optimized | Snowflake, BigQuery, Redshift |
| **Data Lake** | Raw files in any format | S3, Azure Data Lake, GCS |
| **Lakehouse** | Raw storage + warehouse performance | Databricks, Apache Iceberg |

## 1.4 Skills a Data Engineer Needs

| Skill | Why it matters |
|-------|---------------|
| **SQL** | Universal language for querying and transforming data — used every day |
| **Python** | Scripting, pipeline logic, data transformation |
| **Cloud platforms** | AWS, GCP, or Azure — most data infrastructure lives in the cloud |
| **Databases** | Relational (PostgreSQL, MySQL) and columnar (BigQuery, Snowflake) |
| **Orchestration** | Scheduling and monitoring pipelines (Airflow, Prefect, dbt) |
| **Version control** | Git for tracking changes to code and pipeline definitions |

**Soft skills:** breaking down ambiguous requirements into data models; communicating with analysts,
scientists, and business stakeholders; thinking about failure modes.

## 1.5 A Day in the Life

| Time | Activity |
|------|----------|
| **Morning** | Check overnight pipeline runs; investigate failures or data quality alerts |
| **Midday** | Build a new ingestion pipeline; write a SQL transformation model |
| **Afternoon** | Code review; meet with an analyst; update documentation |

The job is roughly 60% building new things and 40% keeping existing things running reliably.

## Key Takeaways

1. Data engineers build **infrastructure** — not insights or predictions directly.
2. Their output is reliable, accessible data that analysts and scientists consume downstream.
3. The role requires SQL, Python, cloud skills, and systems thinking.
4. Without data engineering, neither analytics nor data science can function at scale.
